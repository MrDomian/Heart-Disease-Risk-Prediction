# Profile report
The file generates a comprehensive summary and analysis of the dataset, including statistical measures, data distribution, missing values, correlations and interactive visualisations, using the pandas_profiling library.

### Required libraries

```python
import pandas as pd

from pandas_profiling import ProfileReport
```

### Loading data

```python
data = pd.read_csv('heart_disease_risk.csv')
data
```

### Profile report
ProfileReport creates a profile report for the data contained in the data object. This report contains detailed information about the data, such as descriptive statistics, distributions, correlations, missing values, etc. It generates interactive widgets that allow you to explore the data in a more advanced way, such as filtering, sorting, displaying histograms, etc.

```python
profile = ProfileReport(data)
profile.to_notebook_iframe()
profile.to_widgets()
```
